```python
from openai import OpenAI

client = OpenAI(api_key="xxx")
```

```python
def fetchemail():
    return "Hi"
```

```python
def reader(email_text):

    if "delayed" in email_text:
        return "epilepsy awareness project delayed"

    elif "on-track" in email_text:
        return "epilepsy awareness project on track"

    return "general epilepsy awareness update"
```

```python
def rag(query):

    knowledge = {
        "epilepsy awareness project delayed":
            "Awareness  activities are delayed.",

        "epilepsy awareness project on track":
            "Awareness is now progressing on time.",

        "general epilepsy awareness update":
            "Project update received."
    }

    return knowledge.get(query)
```

```python
def respondert(emailt, summary, context):

    prompt = f"""
    Email:
    {emailt}

    Summary:
    {summary}

    Context:
    {context}

    Write a short professional reply.
    """

    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ]
    )

    return response.choices[0].message.content
```

```python
def reviewert(draft):
    return draft
```

```python
def send_email(to_address, subject, body):


    print( to_address)
    print( subject)
    print( body)
```

```python
mem = []
def run_workflow():

    # email received
    inputemail = fetchemail()

    print(inputemail)

    # summary
    summary = reader(inputemail)

    print(summary)

    # rag
    context = rag(summary)

    print(context)

    # respond
    draft = respondert(inputemail, summary, context)

    print(draft)

    # review
    finalreply = reviewert(draft)

    print(finalreply)

    # memory
    mem.append({
        "email": inputemail,
        "response": finalreply
    })


    print(mem)

    # decision
    choice = input("Send email? (y/n): ")

    if choice == "y":

        send_email(
            "sukhmani@gmail.com",
            "Awareness",
            finalreply
        )

        print(" Email sent")

    else:
        print("Email not sent")
```
