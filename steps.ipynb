```python
import time
```

```python
def fetchemail():
    return "Hi"
```

```python
def reader(email_text):

    if "delayed" in email_text:
        return "project delayed"

    elif "on-track" in email_text:
        return "project on track"

    return "general update"
```

```python
def rag(query):

    knowledge = {
        "project delayed": "Timeline risk detected.",
        "project on track": "Project progressing normally.",
        "general update": "No major concerns."
    }

    return knowledge.get(query)
```

```python
def respondert(email_text, summary, context):

    return (
        "Thank you for your email. "
        "will share updates soon."
    )
```

```python
def reviewert(draft):
    return draft.strip() + " Regards."
```

```python
def send_email(to_address, subject, body):

    print(" EMAIL SENT ")
    print("To:", to_address)
    print("Subject:", subject)
    print("Body:", body)
```

```python
memory = []
def run_workflow():

    # email received
    inputemail = fetchemail()
    print("[raw email]")
    print(inputemail)

    # summary
    summary = reader(inputemail)
    print("[summary]")
    print(summary)

    # rag
    context = rag(summary)
    print("[rag context]")
    print(context)

    # respond
    draft = respondert(inputemail, summary, context)
    print("[DRAFT RESPONSE]")
    print(draft)

    # review
    finalreply = reviewert(draft)
    print("[reviewed reply]")
    print(finalreply)

    # memory
    memory.append({
        "email": inputemail,
        "response": finalreply
    })

    print("[memory]")
    print(memory)

    # decision
    choice = input("Send email? (y/n): ").strip().lower()

    if choice == "y":

        send_email(
            "sukhmani@gmail.com",
            "Re: Project Update",
            finalreply
        )

        print("Status: Email sent")

    else:
        print("Status: Email not sent")
```
