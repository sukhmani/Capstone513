```python
def fetchemail():
    return "Hi"
```

```python
def reader(email_text):
    return "User request."
```

```python
def rag(query):
    return "statusrag."
```

```python
def respondert(email_text, summary, context):

    return (
        "Thank you for your email. "
        "will share updates soon."
    )
```

```python
def reviewert(draft):
    return draft.strip() + " Regards."
```

```python
def send_email(to_address, subject, body):

    print(" EMAIL SENT ")
    print("To:", to_address)
    print("Subject:", subject)
    print("Body:", body)
```

```python
def run_workflow():

    #email received
    inputemail = fetchemail()
    print("[raw email]")
    print(inputemail)

    #summary
    summary = reader(inputemail)
    print("[summary]")
    print(summary)

    #rag
    context = rag(summary)
    print("[rag context]")
    print(context)

    #respond
    draft = respondert(inputemail, summary, context)
    print("[DRAFT RESPONSE]")
    print(draft)

    #review
    finalreply = reviewert(draft)
    print("[reviewed reply]")
    print(finalreply)

    # decision 
    choice = input("Send email? (y/n): ").strip().lower()

    if choice == "y":
        send_email(
            "sukhmani@gmail.com",
            "Re: Project Update",
            finalreply
        )
        print("Status: Email sent")
    else:
        print("Status: Email not sent")
```
